```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
df=pd.read_csv(r'C:/Users/SHIVDAS/Downloads/Titanic.csv')
df.head()
```

```python
df.isna().sum()
```

# Encoding

```python
df=pd.get_dummies(df)
df
```

```python
frequent_items = apriori(df, min_support=0.2, use_colnames=True)
```

```python
frequent_items
```

```python
model = association_rules(frequent_items,metric='lift',min_threshold=0.2)
```

```python
model
```
